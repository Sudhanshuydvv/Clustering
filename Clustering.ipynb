#**Theoretical Questions:**

#1. What is unsupervised learning in the context of machine learning?
 - Unsupervised learning is a type of machine learning where the model is trained on data without labeled responses. Unlike supervised learning, where the algorithm learns from input-output pairs, unsupervised learning explores the structure of tha data to uncover hidden patterns or groupings.

#2. How does K-Means clustering algorithm work?
 - K-Means is one the most popular unsupervised learning  algorithms used for clustering. It partitions a dataset into K distinct, non-overlapping clusters based on feature similarity.

**Step-by-Step Process**
1. **Choose the number of clusters (K)** You decide how many clusters you want the algorithm to find.
2. **Initialize centroids** Randomly select K data points as the initial cluster centroids.
3. **Assign points to the nearest centroid** Each data point is assigned to the cluster whose centroid is closest (usually using Euclidean distance).
4. **Update centroids** Recalculate the centroids as the mean of all points assigned to each cluster.
5. **Repeat steps 3 and 4** Continue until the centroids no longer change significantly or a maximum number of iterations is reached.

#3. Explain the concept of a dendrogram in hierarchical clustering.
 - A dendrogram is a key visualization tool used in hierchical clustering to represent the arrangement of the clusters formed during the algorithm's execution.
 - A dendrogram is a tree-like diagram that records the sequences of merges or splits in hierchical clustering. It helps visualize how clusters are formed step by step and at what level of similarity or distance.

#4. What is the main difference between K-Means and Hierarchical Clustering?
 - Differnces between K-Means and Hierarchichal Clustering:

 1. K-Means -
   - It partitional (divides data into K clusters).
   - It must be specified in advance.
   - It assumes spherical clusters (based on centroids).
   - Efficient for large datasets.

 2. Hierarchichal Clustering -
   - Hierchichal (builds a tree of clusters).
   - Not required; It can be chosen by cutting the dendrogram.
   - It can capture nested and non-spherical structures.
   - It computationally expensive for large datasets.

#5. What are the advantages of DBSCAN over K-Means?
 - Advantages of DBSCAN over K-Means:                    
    - It can find arbitrarily shaped clusters (e.g., moons, spirals).
    - No need to specify the number of clusters beforehand.
    - It works well with clusters of varying densities.
    - No initialization needed.

#9. How does DBSCAN identify noise points?
 - DBSCAN (Density-Based Spatial Clustering of Applications with Noise) identifies noise points based on the density of data points in a region. Here's how it works:

 Core Concepts in DBSCAN

   - Œµ (epsilon): The radius around a point to search for neighbors.
   - minPts: The minimum number of points required within Œµ to form a dense region.

   Noise Detection in Action                                  
Imagine a sparse point sitting far from any dense region:

   - DBSCAN checks its Œµ-neighborhood.
   - If it finds fewer than minPts points, and it‚Äôs not close to any core point, it‚Äôs labeled as noise.

#10. Define inertia in the context of K-Means.
 - Inertia is the sum of squared distances between each data point and the centroid of the cluster it belongs to. It quantifies how tightly the data points are grouped around their centroids.

#11. What is the elbow method in K-Means clustering?
 - The Elbow Method involves plotting the inertia (within-cluster sum of squared distances) against different values of k, and identifying the point where the rate of decrease sharply slows down-this point resembles an "elbow".

#12. Describe the concept of "density" in DBSCAN.
 - In DBSCAN (Density-Based Spatial Clustering of Applications with Noise), density refers to how many data points are packed into a region of space. A region is considered dense if it contains a minimum number of points (minPts) within a given radius (Œµ).

#13. Can hierarchical clustering be used on categorical data?
 - Yes, hierarchical clustering can be used on categorical data, but with some important considerations and adaptations:

**Key Considerations for Categorical Data**          
Hierarchical clustering, especially agglomerative clustering, is flexible in terms of the distance metric used. Since categorical data doesn't have a natural notion of Euclidean distance, you need to:
1. Use Appropriate Distance Measures
  -  Hamming Distance
  - Jaccard Distance
  - Gower Distance

2. Chhose a Suitable Linkage Method:
  - Most linkage methods (e.g., average, complete, single) can be used as long as the distance matrix is precomputed using a suitable metric.

#15. Explain the term "linkage criteria" in hierarchical clustering.
 - In hierarchical clustering, the term "linkage criteria" refers to the method used to determine the distance between clusters when merging them. It plays a crucial role in shaping the structure of the resulting dendrogram and ultimately affects the clustering outcome.

#16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?
 - K-Means clustering can perform poorly on data with varying cluster sizes or densities due to its underlying assumptions and distance-based nature. Here's why:
 1. Assumes Equal Cluster Sizes
 2. Sensitive to Density Differences
 3. Centroid-Based Assignment
 4. Poor Performance on Non-Globular Shapes

#17. What are the core parameters in DBSCAN, and how do they influence clustering?
 - In DBSCAN, the two parameters that control how clusters are formed are:
 1. Œµ (epsilon)
 2. minPts

 **How they influence together -**
   - A point is a **core point** if it has at least `minPts` neighbors within Œµ.
   - Clusters are formed by **connecting core points** and their **density-reachable** neighbors.
   - Points that are not reachable from any core point are labeled as **noise**.

#18. How does K-Means++ improve upon standard K-Means initialization?
 - Here's how it works and why it  matters:

 How K-Means++ Improves Initialization-
   - Choose the first centroid randomly from the data points.
   - for each remaining point, computes the distance squared to the nearest already chosen centroid.
   - Select the next centroid with probability proportational to th distance squared.
   - Repeat until K cetroids are chosen.

   This ensures that:
   - Centroids are spread out across the data space.
   - Initial clusters are more representative, reducing the chance of poor local minima.

#19. What is agglomerative clustering?
 - Agglomerative clustering starts with each data point as its own cluster and then iteratively merges the closest pairs of clusters until:
   - All points belong to a single cluster, or
   - A stopping criterion (like number of clusters or distance threshold) is met.

#20. What makes Silhouette Score a better metric than just inertia for model evaluation?

Definition: Combines both cohesion (how close points are within a cluster) and separation (how far apart clusters are).

Formula for a point

where:
   - ùëé (ùëñ) = average distance to points in the same cluster (intra-cluster distance)

#21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=42)

kmeans = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans.fit_predict(X)

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=200, c='red', marker='X', label='Centroids')
plt.title("K-Means Clustering with 4 Centers")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```

#24. Load the Wine dataset and apply K-Means clustering after standardizing the features. Print the size of each cluster.

#29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

X, _ = make_blobs(n_samples=500, centers=4, cluster_std=0.60, random_state=42)

silhouette_scores = []
k_values = range(2, 6)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)

plt.figure(figsize=(8, 6))
plt.bar(k_values, silhouette_scores, color='skyblue')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for k = 2 to 5")
plt.xticks(k_values)
plt.grid(axis='y')
plt.show()
```

#31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

X, y_true = make_blobs(n_samples=500, centers=3, cluster_std=2.5, random_state=42)

X_scaled = StandardScaler().fit_transform(X)

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)
labels = kmeans.predict(X_scaled)

h = 0.02
x_min, x_max = X_scaled[:, 0].min() - 1, X_scaled[:, 0].max() + 1
y_min, y_max = X_scaled[:, 1].min() - 1, X_scaled[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 7))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, s=50, cmap='viridis', edgecolor='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', s=200, marker='X', label='Centroids')
plt.title("K-Means Clustering with Decision Boundaries (Overlapping Clusters)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```

#34. Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

data = load_breast_cancer()
X = data.data

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

inertias = []
k_values = range(2, 7)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(k_values, inertias, marker='o', linestyle='-', color='teal')
plt.title("K-Means Inertia for K = 2 to 6 (Breast Cancer Dataset)")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.xticks(k_values)
plt.grid(True)
plt.show()
```

#44. Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.preprocessing import StandardScaler
import numpy as np

iris = load_iris()
X = iris.data
X_scaled = StandardScaler().fit_transform(X)

kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X_scaled)

sample_silhouette_values = silhouette_samples(X_scaled, labels)

plt.figure(figsize=(8, 6))
y_ticks = []
y_lower = 0
for i in range(3):
    cluster_values = sample_silhouette_values[labels == i]
    cluster_values.sort()
    y_upper = y_lower + len(cluster_values)
    plt.barh(range(y_lower, y_upper), cluster_values, edgecolor='k')
    y_ticks.append((y_lower + y_upper) / 2)
    y_lower = y_upper

plt.axvline(x=silhouette_score(X_scaled, labels), color="red", linestyle="--", label="Average Silhouette Score")
plt.yticks(y_ticks, [f"Cluster {i}" for i in range(3)])
plt.xlabel("Silhouette Coefficient")
plt.ylabel("Samples")
plt.title("Silhouette Plot for KMeans Clustering (Iris Dataset)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

#47. Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import numpy as np

X, _ = make_blobs(n_samples=500, centers=4, cluster_std=1.2, random_state=42)

dbscan = DBSCAN(eps=1.1, min_samples=5)
labels = dbscan.fit_predict(X)

n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = np.sum(labels == -1)

print(f"Number of clusters found: {n_clusters}")
print(f"Number of noise points: {n_noise}")
```

#48. Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering

digits = load_digits()
X = digits.data
X_scaled = StandardScaler().fit_transform(X)

tsne = TSNE(n_components=2, perplexity=30, learning_rate=200, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)

agglo = AgglomerativeClustering(n_clusters=10, linkage='ward')
labels = agglo.fit_predict(X_tsne)

plt.figure(figsize=(10, 7))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels, cmap='tab10', s=50, alpha=0.7, edgecolors='k')
plt.title("Agglomerative Clustering on Digits Dataset (2D t-SNE Projection)")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.grid(True)
plt.show()
```
